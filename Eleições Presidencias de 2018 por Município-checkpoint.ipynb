```python
# Vamos carregar o arquivo com o resultado da eleição no Brasil dividido por municipio/zona
dados_votacao = pd.read_csv('C:/Users/SilvioOliver/Documents/eleicao/votacao/votacao_candidato_munzona_2018_BR.csv',
                        sep=';', encoding='latin1')

dados_votacao.head()
```

```python
# Visualizando o dataframe transposto
dados_votacao.head().T
```

```python
#  mostrar duas colunas, uma com o índice e a outra com todos os dados da coluna, ou seja, 1 ou 2, 1 correspondendo ao primeiro turno e 2 ao segundo turno
dados_votacao['NR_TURNO']
```

```python
# Separar os dados pois precisamos apenas dos dados de 2 Turno
# dados_votacao[dados_votacao['NR_TURNO'] == 1] caso for  1 turno
dados_votacao = dados_votacao[dados_votacao['NR_TURNO'] == 2]

dados_votacao.sample(3).T
```

```python
# Vamos verificar quantos municipios unicos existem no conjunto de dados
# Vamos pegar por codigo do municipio
dados_votacao['CD_MUNICIPIO'].nunique()
```

```python
# Observem o 'ZZ' acima, corresponde aos eleitores de fora do país
# Retirando os eleitores de fora do país
# Quantas cidades fora do Brasil estão no conjunto de dados?
# Mas o Brasil tem 5570 municipios
# Vamos verificar algumas cidades fora do Brasil
dados_votacao[dados_votacao['SG_UF'] == 'ZZ']['CD_MUNICIPIO'].nunique()
```

```python
# Vamos eliminar os municpios de fora do Brasil
# codigo são votações no exterior
dados_votacao = dados_votacao[dados_votacao['SG_UF'] != 'ZZ']
```

```python
# Verificando a quantidade de municpios
# Contabilizando a quantidade de municípios
dados_votacao['CD_MUNICIPIO'].nunique()
```

```python
# Os dados estão separados por municipio/zona
# Municipios pequenos normalmente tem apenas 1 zona eleitoral
# mas municipios maiores podem ter várias zonas eleitorias

# Vamos verificar Recife
dados_votacao[dados_votacao['NM_MUNICIPIO'] == 'RECIFE']
```

```python
# Eliminar algumas colunas do dataframe que não serão utilizadas
dados_votacao = dados_votacao[['SG_UF', 'CD_MUNICIPIO', 'NM_MUNICIPIO', 'NR_CANDIDATO',
                        'NM_URNA_CANDIDATO', 'QT_VOTOS_NOMINAIS']]

dados_votacao.head()
```

```python
# Nome dos candidatos 
dados_votacao['NM_URNA_CANDIDATO'].unique()
```

```python
# Vamos dar uma investigada vendo a quantidade de valores únicos de unidade da federação
dados_votacao['SG_UF'].unique()
```

```python
len(dados_votacao[(dados_votacao['NM_MUNICIPIO'] == 'RECIFE') & (dados_votacao['NM_URNA_CANDIDATO'] == 'JAIR BOLSONARO')])
```

```python
# Precisamos agregar os dados por municipio e candidato
# Percebemos que só o candidato A apresentou 11 linhas.
# Isso acontece devido ao fato da existência de mais de uma zona eleitoral em Recife
# Em pequenos municípios há somente uma zona eleitoral, no máximo duas
# Em municípios maiores, há mais de uma zona eleitoral, então precisamos agrupar os resultados 
# para que por município visualizemos a quantidade de linhas de candidatos que obtiveram voto naquele município.
# Para isso, iremos agrupar por UF, Código de município, Nome de município e nome de candidato, e iremos somar os votos.
dados_votacao.groupby(['SG_UF', 'CD_MUNICIPIO', 'NM_MUNICIPIO', 'NR_CANDIDATO', 'NM_URNA_CANDIDATO']).sum()
```

```python
# Vemos que conseguimos agrupar e somar a quantidade de votos que cada candidato teve por município
# Mas a mudança acima não foi inplace, o dataframe ainda está do mesmo jeito de antes
# Vamos agora modificar o dataframe para que fique agrupado como acima e resetar o índice, pois perdemos o índice no groupby

# Fazer o groupby, reindexar o indice, vai salvar isso em um dataframe
dados_votacao = dados_votacao.groupby(['SG_UF', 'CD_MUNICIPIO', 'NM_MUNICIPIO', 'NR_CANDIDATO', 'NM_URNA_CANDIDATO']).sum().reset_index()

dados_votacao.sample(4)
```

```python
# Vamos verificar novamente a cidade de Recife
dados_votacao[dados_votacao['NM_MUNICIPIO'] == 'RECIFE']
```

```python
# Percebemos que agora cada município só tem a quantidade de linhas de candidatos que receberam votos
# Agora preciso deixar somente uma linha pra cada município, para ver o candidato vencedor naquele município
# Tem várias formas de fazer isso, mas uma das que parece mais simples é ordenar pelo mais votado e deletar os outros do dataframe

# Ordenar o dataframe pela quantidade de votos e depois deixar apenas o primeiro
# registro de cada municipio
dados_votacao.sort_values(by='QT_VOTOS_NOMINAIS', ascending=False).head(10)

# dropar as linhas duplicadas, mantendo a primeira
dados_votacao.drop_duplicates(subset='CD_MUNICIPIO', keep='first', inplace=True)

dados_votacao.head(10)
```

```python
# Vamos verificar novamente a cidade de Recife
dados_votacao[dados_votacao['NM_MUNICIPIO'] == 'RECIFE']
```

```python
# Agora vamos criar uma coluna representando o resultado da eleição naquele municipio
# sendo que o Resultado é True se o candidato Bolsonaro venceu naquele municipio
# e False em caso contrário
dados_votacao['RESULTADO'] = dados_votacao['NR_CANDIDATO'].apply(lambda x: x == 17)
```

```python
# Recuperar a tabela de correspondencia entre IBGE e TSE
# direto do github do Estadão
link = 'https://raw.githubusercontent.com/estadao/como-votou-sua-vizinhanca/master/data/votos/correspondencia-tse-ibge.csv'

dados_equivalencia = pd.read_csv(link)

dados_equivalencia.head()
```

```python
# Vamos preparar os dataframes
```

```python
# Vamos reindexar o dataframe de votação pelo código do TSE
dados_vot_novo = dados_votacao.copy()

dados_vot_novo.set_index('CD_MUNICIPIO', drop=False, inplace=True)

dados_vot_novo.head()
```

```python
# Vamos reindexar o dataframe de Equivalencia pelo Código do TSE
dados_equi_novo = dados_equivalencia.copy()

dados_equi_novo.set_index('COD_TSE', drop=False, inplace=True)

dados_equi_novo.head()
```

```python
# Verificando o tamanho dos dataframes
dados_vot_novo.info()
```

```python
dados_equi_novo.info()
```

```python
# Criando um novo dataframe com base no merge de Votação e Equivalencia
dados_vot_equi = dados_vot_novo.join(dados_equi_novo)

dados_vot_equi.head()
```

```python
# Vamos juntar o dataframe do GeoPandas com o dataframe Votação/Equivalência
```

```python
dados_vot_equi.info()
```

```python
dados_mapa.info()
```

```python
# Transformar a coluna de codigo de IBGE do dataframe de Votação/Equivalência
dados_vot_equi['GEOCOD_IBGE'] = dados_vot_equi['GEOCOD_IBGE'].astype(str)

dados_vot_equi.info()
```

```python
# Vamos unificar os indices

# Mudando o indice do dataframe de Mapa
dados_mapa.set_index('CD_GEOCMU', drop=False, inplace=True)
```

```python
# Mudando o indice do dataframe de Votação/Equivalência
dados_vot_equi.set_index('GEOCOD_IBGE', drop=False, inplace=True)
```

```python
# Vamos juntar os dataframes
dados_mapa_novo = dados_mapa.join(dados_vot_equi)

dados_mapa_novo.head().T
```

```python
# Verificando o tamanho do mapa novo
dados_mapa_novo.info()
```

```python
# Vamos descobrir os 2 municipios a mais no mapa
dados_mapa_novo[dados_mapa_novo['NR_CANDIDATO'].isnull()]
```

```python
# Matar a coluna ajuste
dados_mapa_novo.drop(columns=['AJUSTE'], inplace=True)
```

```python
# Vamos eliminar a informação das duas lagoas do RS
dados_mapa_novo.dropna(inplace=True)
```

```python
dados_mapa_novo.info()
```

```python

# Vamos desenhar o mapa

# Criar a referência para o código de cor (False/0 = red, True/1 = green)
from matplotlib.colors import ListedColormap

cmap = ListedColormap(['red', 'green'])

# Desenhando o mapa
fig, ax = plt.subplots(1, figsize=(12, 12))

ax = dados_mapa_novo.plot(column='RESULTADO', cmap=cmap, legend=True, linewidth=0.1, ax=ax,
                      edgecolor='grey')

# Escondendo o eixos
ax.set_axis_off()

# Colocando um título
ax.set_title('Resultado Eleição 2018 por Município / 2º Turno', fontdict={'fontsize': '25', 'fontweight': '3'})

# Criando uma nota do rodapé
ax.annotate('Fonte: IBGE, 2015; TSE, 2019; Jornal o Estado de São Paulo, 2019',
    xy=(0.1, 0.08), xycoords='figure fraction', horizontalalignment='left',
    verticalalignment='top', fontsize=10, color='#555555')

# Alterando a legenda
leg = ax.get_legend()
leg.get_texts()[0].set_text('Fernando Haddad')
leg.get_texts()[1].set_text('Jair Bolsonaro')

plt.show()
```
